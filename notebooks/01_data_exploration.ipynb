#### Title & Objective

#### Data Exploration & Synthetic Dataset Generation

**Project:** Queens Supermarket Sales ETL Pipeline  
**Client Context:** Midroc Commerce (Queens Supermarket - Carrefour Partner)  

**Objective:**  
Generate a realistic, synthetic supermarket sales dataset that mimics real-world operations of Queens Supermarket in Ethiopia. This dataset will be used throughout the ETL pipeline to simulate ERP/POS data coming from multiple branches.

**Key Features:**
- Multiple branches in Addis Ababa and other cities
- Different product lines relevant to Ethiopian retail market
- Realistic customer behavior and payment methods
- Time series data spanning 12 months
- Business metrics (Tax, Gross Income, COGS, Margin)

#### Import Libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import os

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Libraries imported successfully!")
```

#### Define Parameters & Configuration

```python
branches = [
    {"branch_id": "Q001", "branch_name": "Queens - Bole",      "city": "Addis Ababa"},
    {"branch_id": "Q002", "branch_name": "Queens - CMC",       "city": "Addis Ababa"},
    {"branch_id": "Q003", "branch_name": "Queens - Piassa",    "city": "Addis Ababa"},
    {"branch_id": "Q004", "branch_name": "Queens - Hawassa",   "city": "Hawassa"},
    {"branch_id": "Q005", "branch_name": "Queens - Adama",     "city": "Adama"}
]

# Product Lines relevant to Ethiopian supermarket market
product_lines = [
    "Food and Beverages", 
    "Household and Cleaning", 
    "Health and Beauty", 
    "Fashion Accessories",
    "Electronic Accessories", 
    "Snacks and Confectionery",
    "Fresh Produce", 
    "Beverages"
]

# Payment Methods common in Ethiopia
payment_methods = ["Cash", "Telebirr", "CBE Birr", "Card"]

# Customer Types
customer_types = ["Member", "Normal"]

# Time period: Full year 2025
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)

# Number of transactions
n_transactions = 50_000

print("Realistic business parameters defined successfully!")
```

#### Generate Synthetic Dataset

```python
# ==================== SYNTHETIC DATA GENERATION (Realistic Version) ====================

data = []

for i in range(n_transactions):
    # Random date and time (supermarket operating hours: 8 AM - 10 PM)
    days_delta = random.randint(0, (end_date - start_date).days)
    transaction_date = start_date + timedelta(days=days_delta)
    hour = random.randint(8, 21)
    minute = random.randint(0, 59)
    transaction_time = f"{hour:02d}:{minute:02d}"
    
    # Select branch
    branch = random.choice(branches)
    
    # Product details - More realistic pricing
    product_line = random.choice(product_lines)
    
    # Different price ranges based on product category
    if product_line in ["Fresh Produce", "Snacks and Confectionery"]:
        unit_price = round(random.uniform(25, 180), 2)
    elif product_line in ["Food and Beverages", "Household and Cleaning"]:
        unit_price = round(random.uniform(40, 450), 2)
    elif product_line == "Health and Beauty":
        unit_price = round(random.uniform(80, 850), 2)
    else:
        unit_price = round(random.uniform(120, 1200), 2)
    
    quantity = random.randint(1, 12)         
    
    # Calculations
    subtotal = round(unit_price * quantity, 2)
    tax_15 = round(subtotal * 0.15, 2)        
    total = round(subtotal + tax_15, 2)
    
    # COGS and Margin
    cogs = round(subtotal * random.uniform(0.62, 0.78), 2)
    gross_income = round(total - cogs, 2)
    gross_margin_percentage = round((gross_income / total) * 100, 2)
    
    # Customer & Payment
    customer_type = random.choice(customer_types)
    gender = random.choice(["Male", "Female"])
    payment = random.choice(payment_methods)
    rating = round(random.uniform(4.0, 9.5), 1)
    
    data.append({
        "invoice_id": f"INV{str(100000 + i).zfill(6)}",
        "branch_id": branch["branch_id"],
        "branch_name": branch["branch_name"],
        "city": branch["city"],
        "customer_type": customer_type,
        "gender": gender,
        "product_line": product_line,
        "unit_price": unit_price,
        "quantity": quantity,
        "tax_15": tax_15,
        "total": total,
        "date": transaction_date.strftime("%Y-%m-%d"),
        "time": transaction_time,
        "payment": payment,
        "cogs": cogs,
        "gross_income": gross_income,
        "gross_margin_percentage": gross_margin_percentage,
        "rating": rating
    })

# Create DataFrame
df = pd.DataFrame(data)

# Final check
print(f"Realistic synthetic dataset generated successfully!")
print(f"Total Transactions : {len(df):,}")
print(f"Date Range         : {df['date'].min()} to {df['date'].max()}")
print(f"Average Transaction: {df['total'].mean():,.2f} ETB")
print(f"Average Quantity   : {df['quantity'].mean():.2f} items")
print(f"Average Unit Price : {df['unit_price'].mean():,.2f} ETB")
```

#### Exploratory Analysis (Business Insights)

##### 1. Overall Sales Summary

```python
print("="*70)
print("OVERALL SALES SUMMARY - QUEENS SUPERMARKET (2025)")
print("="*70)
total_sales = df['total'].sum()
total_transactions = len(df)

print(f"Total Revenue (2025)          : {total_sales:,.2f} ETB")
print(f"Total Transactions            : {total_transactions:,}")
print(f"Average Transaction Value     : {df['total'].mean():,.2f} ETB")
print(f"Average Items per Basket      : {df['quantity'].mean():.2f}")
print(f"Average Gross Margin          : {df['gross_margin_percentage'].mean():.2f}%")
print(f"Total Gross Income           : {df['gross_income'].sum():,.2f} ETB\n")
```

##### 2. Performance by Branch

```python
print("="*70)
print("BRANCH PERFORMANCE ANALYSIS")
print("="*70)
branch_analysis = df.groupby('branch_name').agg(
    total_sales=('total', 'sum'),
    num_transactions=('invoice_id', 'count'),
    avg_transaction=('total', 'mean'),
    avg_margin=('gross_margin_percentage', 'mean')
).round(2)

branch_analysis = branch_analysis.sort_values(by='total_sales', ascending=False)
print(branch_analysis)
```

##### 3. Top Performing Product Lines

```python
print("\n" + "="*70)
print("PRODUCT LINE PERFORMANCE")
print("="*70)
product_analysis = df.groupby('product_line').agg(
    total_revenue=('total', 'sum'),
    units_sold=('quantity', 'sum'),
    avg_unit_price=('unit_price', 'mean'),
    avg_margin=('gross_margin_percentage', 'mean')
).round(2)

product_analysis = product_analysis.sort_values(by='total_revenue', ascending=False)
print(product_analysis)
```

##### 4. Customer Type & Payment Method Analysis

```python
# 4. Customer & Payment Insights
print("\n" + "="*70)
print("CUSTOMER BEHAVIOR & PAYMENT PREFERENCES")
print("="*70)

print("Sales by Customer Type:")
customer_analysis = df.groupby('customer_type').agg(
    total_sales=('total', 'sum'),
    transactions=('invoice_id', 'count'),
    avg_basket=('total', 'mean')
).round(2)
print(customer_analysis)

print("\nPayment Method Distribution (%):")
print((df['payment'].value_counts(normalize=True) * 100).round(2))
```
